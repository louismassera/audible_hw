## Building the dataset

The goal is to build a dataset that has one row per movie, with the following features:
    - tconst: alphanumeric unique identifier of the title (str)
    - startYear: represents the release year of a title (int)
    - directors: director(s) of the given title (list)
    - writers: writer(s) of the given title (list)
    - actors: actor(s) of the given title (list)
    - genres: includes up to three genres associated with the title (list)
    - numVotes: number of votes the title has received (int)
    - runtimeMinutes: primary runtime of the title, in minutes (int)
    - averageRating: weighted average of all the individual user ratings (float)

## Preprocessing and Imputation

Now that the dataset is ready, we still need some more steps to predict the averageRating. 

### Numerical Features

To fill the missing values of runtimeMinutes and startYear, we use the mean of those features.

```python
df['runtimeMinutes'] = df['runtimeMinutes'].apply(lambda x: np.nan if x == '\\N' else int(x))
avg = math.ceil(df['runtimeMinutes'].mean())
df['runtimeMinutes'] = df['runtimeMinutes'].apply(lambda x: avg if math.isnan(x) else x)
```

```python
df['startYear'] = df['startYear'].apply(lambda x: np.nan if x == '\\N' else int(x))
avg = math.ceil(df['startYear'].mean())
df['startYear'] = df['startYear'].apply(lambda x: avg if math.isnan(x) else x)
```

### Categorical Features

Probably the core of this HW. How to deal with the categorical features here? This is of first importance as most of the information about the quality of the movie is probably there (genres, directors, writers and actors). However, a quick look at the dataset shows that they are way too many directors, writers and actors to One-hot encode them.

#### Genre

However when it comes to genres, there is only a limited amount of them (~20). Hence we can get the dummy variables to analyse them. We will select the top genres during Data Analysis.

```python
dummies_genre = df['genres'].str.join('|').str.get_dummies()
dummies_genre['tconst'] = df['tconst']
df = df.merge(dummies_genre, on='tconst')
```

#### Directors, writers and actors

A lot of directors, writers and actors! About these categories, I could think of two interesting features for each movie. For instance about the directors:
 - rank_directors: chronological position of this movie in the director's filmography (1 being the first one directed). If more than one director was involved, we take the mean.
 - nbr_movie_category: total number of movies directed by this director. Once again, if more than one director was involved, we take the mean.
<br>

Writers and actors features are defined the same way.
The main idea behind that feature creation is to get more interpretable features and reduce the cardinality of the original features (directors regrouping 90K different directors for instance, but they made from 1 to 199 movies). We guess that experience (from the director, the writer and the actors) might be positively correlated with the rating.

```python
def build_ranking(df, category):
    
    col1 = 'rank_' + category
    col2 = 'nbr_movie_' + category
    df_category = pd.DataFrame({'tconst':np.repeat(df.tconst.values, df[category].str.len()),
                                'startYear':np.repeat(df.startYear.values, df[category].str.len()),
                                 category:np.concatenate(df[category].values)})
    df_category = df_category[(df_category.startYear != '\\N') & (df_category[category] != '\\N')]
    
    df_category[col1] = df_category.groupby(by=category)['startYear'].rank(method='first').apply(lambda x: int(x))
    temp = df_category.groupby(by=category)[col1].max().reset_index().rename(index=str, columns={category: category, col1: col2})
    df_category = df_category.merge(temp, on=category)
    rank_avg = df_category.groupby(by='tconst')[col1].mean().apply(lambda x: math.ceil(x)).reset_index()
    nbr_movie_avg = df_category.groupby(by='tconst')[col2].mean().apply(lambda x: math.ceil(x)).reset_index()
    
    return rank_avg, nbr_movie_avg
```

```python
directors_avg, d_nbr_movie_avg = build_ranking(df, 'directors')
writers_avg, w_nbr_movie_avg = build_ranking(df, 'writers')
actors_avg, a_nbr_movie_avg = build_ranking(df, 'actors')
```

```python
df = df.merge(directors_avg, how='left', on='tconst').merge(writers_avg, how='left', on='tconst').merge(actors_avg, how='left', on='tconst')
df = df.merge(d_nbr_movie_avg, how='left', on='tconst').merge(w_nbr_movie_avg, how='left', on='tconst').merge(a_nbr_movie_avg, how='left', on='tconst')
```

```python
for i in ['rank_directors', 'nbr_movie_directors', 'rank_writers', 'nbr_movie_writers', 'rank_actors', 'nbr_movie_actors']:
    df[i] = df[i].apply(lambda x: 1 if math.isnan(x) else x)
```

We can also consider the ratio of these two, representing how far the artist was in his career.  

```python
df['career_directors'] = df['rank_directors']/df['nbr_movie_directors']
df['career_writers'] = df['rank_writers']/df['nbr_movie_writers']
df['career_actors'] = df['rank_actors']/df['nbr_movie_actors']
```

## Model Selection

Now that we have a better idea of what is in the data, we can start the model selection.

### Linear Models

```python
def plot_important_features(coef, feature_names, top_n=20, ax=None, rotation=60):

    if ax is None:
        ax = plt.gca()
    inds = np.argsort(coef)
    low = inds[:top_n]
    high = inds[-top_n:]
    important = np.hstack([low, high])
    myrange = range(len(important))
    colors = ['red'] * top_n + ['blue'] * top_n
    
    ax.bar(myrange, coef[important], color=colors)
    ax.set_xticks(myrange)
    ax.set_xticklabels(feature_names[important], rotation=rotation,
                       ha="right", fontsize=14)
    ax.set_xlim(-.7, 2 * top_n)
    ax.set_frame_on(False)
```

```python
X_train = X_train.drop('target', axis=1)
```

```python
ols_pipe = make_pipeline(MaxAbsScaler(), LinearRegression())
```

```python
selected_features = selected_genre + selected_numerical
```

```python
np.mean(cross_val_score(ols_pipe, 
                        X_train, 
                        y_train, 
                        cv=3))
```

```python
np.mean(cross_val_score(ols_pipe, 
                        X_train[selected_features], 
                        y_train, 
                        cv=3))
```

There is a drop of 4% in the explained variance when using the selected_features, not too bad. Let's select the feature using Lasso regression.

```python
lasso_pipe = make_pipeline(MaxAbsScaler(), Lasso(max_iter=100000))
param_grid = {'lasso__alpha': np.logspace(-5, 3, 9)}
grid = GridSearchCV(lasso_pipe, param_grid, cv=3)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.best_score_)
```

```python
plt.figure(figsize=(15, 6))
plot_important_features(grid.best_estimator_.named_steps['lasso'].coef_,
                        X_train.columns)
plt.show()
```

```python
model = SelectFromModel(grid.best_estimator_.named_steps['lasso'], threshold=-np.inf, max_features=15, prefit=True)
X_new = model.transform(X_train)
```

```python
suport = X_train.columns[model.get_support()]
```

### Evaluation on the test set

```python
final_boost = make_pipeline(MaxAbsScaler(), GradientBoostingRegressor(n_estimators=300, max_features='auto'))
final_boost.fit(X_train, y_train)
final_boost.score(X_test, y_test)
```

```python
y_pred = final_boost.predict(X_test)
R2 = final_boost.score(X_test, y_test)
MAE = mean_absolute_error(y_pred, y_test)
MSE = mean_squared_error(y_pred, y_test)
print('R2 on test set:', R2)
print('MAE on test set:', MAE)
print('MSE on test set:', MSE)
```

Almost 30% of variance explained, which is far from good but better than nothing. The MAE is also encouraging.

## To go further

- Expand the rating space in order to fit the reality. For a movie, going from an average of 9 to 10 is not the same as going from an average of 6 to 7. We need to represent that non linearity. For instance by looking at y**3.

```python
final_boost_expand = make_pipeline(MaxAbsScaler(), GradientBoostingRegressor(n_estimators=300, max_features='auto'))
final_boost_expand.fit(X_train, y_train**3)
y_pred = final_boost_expand.predict(X_test)
R2 = final_boost_expand.score(X_test, y_test**3)
MAE = mean_absolute_error(y_pred**(1/3), y_test)
MSE = mean_squared_error(y_pred**(1/3), y_test)
print('R2 on test set:', R2)
print('MAE on test set:', MAE)
print('MSE on test set:', MSE)
```

It seems to make sense.

- Use target encoding to deal with high cardinality in directors, writers and actors.

```python
target_col = []
tgt = ['rank_directors', 'nbr_movie_directors', 'rank_writers', 'nbr_movie_writers', 'rank_actors', 'nbr_movie_actors']
for j in tgt:
    X_train[j +'_tgt'] = X_train[j].apply(lambda x: 1 if x==1 else (2 if x<100 else 3))
    target_col.append(j +'_tgt')
```

```python
all_tgt = list(set(X_train.columns) - set(target_col) - set(tgt)) 
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ('num', MaxAbsScaler(), all_tgt),
        ('cat', ce.TargetEncoder(cols=target_col), target_col)]
)
model = make_pipeline(preprocess, GradientBoostingRegressor(n_estimators = 300, random_state = 0))
param_grid = {
    'gradientboostingregressor__max_features': ['log2','auto']
}
grid = GridSearchCV(model, param_grid, cv=3, n_jobs=-1, verbose=True)
```

```python
%%time
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.best_score_)
```

## Conclusion

Predicting the average rating is more complex than it seems. Indeed, the numerical variables are not that informative and handling the categorical variable is tricky here (except for genres). I feel target encoding could be useful for the director/writer/actor representation but they have to be regrouped in bigger categories first (a lot of movie director/writer/actor only appear once). What's more, the target has a low variance, making the models aiming for the mean to minimize its risk. Inflating the rating space seemed to help. My best model could explain 30% of the variance, there is a huge scope for improvement. Mixing these datasets with additional sources would greatly help to improve the prediction.
